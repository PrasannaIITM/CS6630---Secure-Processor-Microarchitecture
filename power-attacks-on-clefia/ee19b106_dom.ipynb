```python
def intermediate_f0_HW(x, b, RK):
    xb = x[b] ^ RK

    if b == 0:  
        return bin(int(T0_F0[xb])).count("1")
    elif b == 1:
        return bin(int(T1_F0[xb])).count("1")
    elif b == 2:
        return bin(int(T2_F0[xb])).count("1")
    elif b == 3:
        return bin(int(T3_F0[xb])).count("1")

def intermediate_f1_HW(x, b, RK):    
    xb = x[b] ^ RK

    if b == 0:  
        return bin(int(T0_F1[xb])).count("1")
    elif b == 1:
        return bin(int(T1_F1[xb])).count("1")
    elif b == 2:
        return bin(int(T2_F1[xb])).count("1")
    elif b == 3:
        return bin(int(T3_F1[xb])).count("1")

def f0(x, RK):
    x = [x[i] ^ RK[i] for i in range(len(RK))]
    fout = T0_F0[x[0]] ^ T1_F0[x[1]] ^ T2_F0[x[2]] ^ T3_F0[x[3]]
    return [int(fout & 0xFF), int((fout >> 8) & 0xFF), int((fout >> 16) & 0xFF), int((fout >> 24) & 0xFF)]

def f1(x, RK):
    x = [x[i] ^ RK[i] for i in range(len(RK))]
    fout = T0_F1[x[0]] ^ T1_F1[x[1]] ^ T2_F1[x[2]] ^ T3_F1[x[3]]
    return [int(fout & 0xFF), int((fout >> 8) & 0xFF), int((fout >> 16) & 0xFF), int((fout >> 24) & 0xFF)]
```
